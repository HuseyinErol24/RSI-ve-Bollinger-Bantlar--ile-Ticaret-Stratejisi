# Projenin Amacı

<p>RSI'nin ve Bollinger bantlarının kombinasyonunu kullanarak piyasalardaki potansiyel alım ve satım noktalarını belirlemeyi amaçlıyorum. Bu strateji, piyasadaki trendleri belirleme ve fiyat hareketlerini daha doğru bir şekilde analiz etme konusunda yatırımcılara yardımcı olabilir. Ayrıca, bu sinyalleri geriye dönük olarak test ederek stratejinin geçmiş performansını değerlendirdim ve gerçek piyasalarda da etkinliğini test etmeye çalıştım.</p>

<p>Projede kullandığım göstergelerden biri olan <strong>RSI (Relative Strength Index)</strong>, bir varlığın aşırı alım veya aşırı satım durumlarını belirlemek için kullanılan bir momentum osilatörüdür. <strong>Bollinger bantları</strong> ise fiyat hareketlerinin volatilitesini gösteren bir göstergedir. Bu göstergelerin kombinasyonunu kullanarak,piyasada mevcut fırsatları daha iyi değerlendirebilir ve daha bilinçli yatırım kararları alabiliriz. Bu sayede, yatırımcılar finansal başarılarını artırabilir ve portföylerini daha etkin bir şekilde yönetebilirler.</p>

<h3>Kodlama kısmından önce stratejimin daha net anlaşılması adına TradingView kullanarak stratejimi anlatmaya çalışacağım</h3>

<h1 style="text-align: center; color: green; font-size: 24px;">Alış Stratejisi</h1>


<h1>Strateji Adımları</h1>
    
<h2>Sinyal Başlangıcı (İlk Mum)</h2>
    <ul>
        <li><strong>Koşul:</strong> Hisse senedinin kapanış fiyatı, Bollinger bantlarının alt bandının altında kalırsa.</li>
        <li><strong>Açıklama:</strong> Bu durum, fiyatın aşırı satım bölgesinde olabileceğini ve potansiyel bir alım fırsatının oluşabileceğini gösterir. İlk mum, bu sinyalin başladığını gösterir.</li>
    </ul>

<h2>Kontrol Mumu (İkinci Mum)</h2>
    <ul>
        <li><strong>Koşul:</strong> İkinci mum, hareketi doğrular. İkinci mumun kapanış fiyatı, alt bandın altında kalır.</li>
        <li><strong>Açıklama:</strong> İlk mumdan sonra gelen bu ikinci mum, fiyatın hala aşırı satım bölgesinde olduğunu ve sinyalin geçerli olduğunu teyit eder. Bu mum, stratejideki teyit mumudur.</li>
    </ul>

<h2>RSI Kontrolü</h2>
    <ul>
        <li><strong>Koşul:</strong> RSI (Relative Strength Index) değeri artıyorsa.</li>
        <li><strong>Açıklama:</strong> RSI, fiyatın aşırı alım veya aşırı satım bölgesinde olup olmadığını gösteren bir momentum osilatörüdür. RSI'ın artması, fiyatın toparlanmaya başladığını ve alım sinyalinin güçlendiğini gösterir.</li>
    </ul>

<h1>Strateji Şartları</h1>
    
<h2>Alt Bandın Altında Kapanış</h2>
    <ul>
        <li>İlk mum, kapanış fiyatının Bollinger bantlarının alt bandının altında olmasıyla sinyal verir.</li>
        <li>İkinci mum da kapanış fiyatının Bollinger bantlarının alt bandının altında olmasıyla kontrol edilir.</li>
    </ul>

<h2>RSI Değeri</h2>
    <ul>
        <li>RSI değeri, ikinci mumu takiben artmaya başlamalıdır.</li>
    </ul>

<h1 style="text-align: center; color: red; font-size: 24px;">Satış Stratejisi</h1>


<h2>Strateji Adımları</h2>
    
<h3>Sinyal Başlangıcı (İlk Mum)</h3>
    <ul>
        <li><strong>Koşul:</strong> Hisse senedinin kapanış fiyatı, Bollinger bantlarının üst bandının üzerinde kalırsa.</li>
        <li><strong>Açıklama:</strong> Bu durum, fiyatın aşırı alım bölgesinde olabileceğini ve potansiyel bir satış fırsatının oluşabileceğini gösterir. İlk mum, bu sinyalin başladığını gösterir.</li>
    </ul>

<h3>Kontrol Mumu (İkinci Mum)</h3>
    <ul>
        <li><strong>Koşul:</strong> İkinci mumun kapanış fiyatı, üst bandın üzerinde kalır.</li>
        <li><strong>Açıklama:</strong> İlk mumdan sonra gelen bu ikinci mum, fiyatın hala aşırı alım bölgesinde olduğunu ve sinyalin geçerli olduğunu teyit eder.</li>
    </ul>

<h3>RSI Kontrolü</h3>
    <ul>
        <li><strong>Koşul:</strong> RSI (Relative Strength Index) değeri azalıyorsa.</li>
        <li><strong>Açıklama:</strong> RSI, fiyatın aşırı alım bölgesinde olup olmadığını gösteren bir momentum osilatörüdür. RSI'ın azalması, fiyatın düşmeye başladığını ve satış sinyalinin güçlendiğini gösterir.</li>
    </ul>
<h2>Strateji Şartları</h2>
    
<h3>Üst Bandın Üstünde Kapanış</h3>
    <ul>
        <li>İlk mumun kapanış fiyatı, Bollinger bantlarının üst bandının üzerinde olursa.</li>
        <li>İkinci mumun kapanış fiyatı da üst bandın üzerinde olursa.</li>
    </ul>

<h3>RSI Değeri</h3>
    <ul>
        <li>RSI değeri, ikinci mumun RSI değerinden büyük olmalıdır.</li>
    </ul>

```python
#gerekli kütüphaneler import edilir
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go
```

```python
veri = yf.download("KCHOL.IS", start="2011-05-01") 
# yf kullanarak veriler indirildi
```

```python
plt.figure(figsize=(15,6))
plt.title("KCHOL Kapanış Verileri")
plt.plot(veri["Close"],label="Kapanış Değeri")
plt.legend()
plt.grid()
```

# Bolinger Bantlarının hesaplanması

```python
orta_bant_periot = 20
Orta_bant = veri["Close"].rolling(window=orta_bant_periot).mean()
Üst_bant_periot = 20
ortalama = veri["Close"].rolling(window=Üst_bant_periot).mean()
standart_sapma = veri["Close"].rolling(window=Üst_bant_periot).std()
Üst_bant = ortalama + (standart_sapma * 2)
alt_bant = 20
ortalama = veri["Close"].rolling(window=alt_bant).mean()
standart_sapma = veri["Close"].rolling(window=alt_bant).std()
alt_bant = ortalama - (standart_sapma * 2)
veri["Üst_bant"] = np.array(Üst_bant) 
veri["Orta_bant"] = np.array(Orta_bant)
veri["alt_bant"] = np.array(alt_bant)
#orta alt ve üst bantlar hesaplanarak dateframe aktarlılır
```

<br><br><br><br><br><br><br><br><br><br><br><br><br><br>

# mum grafik üzerinde bolinger bantları

```python
#verileri bu şekilde göstermek hem stratejiminn daha iyi anlaşılmasını saglarken hemde daha detaylı şekilde
#bolinger bantları ve hareket arasındaki ilişkiyi  anlamamızı saglar 
fig = go.Figure(data=[go.Candlestick(
    x=veri.index,
    open=veri['Open'],
    high=veri['High'],
    low=veri['Low'],
    close=veri['Close'],
    name='Kapanış Fiyatı'
)])

fig.add_trace(go.Scatter(
    x=veri.index, 
    y=veri['Üst_bant'], 
    line=dict(color='blue', width=1), 
    name='Üst Bant'
))

fig.add_trace(go.Scatter(
    x=veri.index, 
    y=veri['Orta_bant'], 
    line=dict(color='red', width=1), 
    name='Orta Bant'
))

fig.add_trace(go.Scatter(
    x=veri.index, 
    y=veri['alt_bant'], 
    line=dict(color='blue', width=1), 
    name='Alt Bant'
))

fig.update_layout(
    title="KCHOL Hisse Senedi Kapanış Fiyatı ve Bollinger Bantları",
    xaxis_title='Tarih',
    yaxis_title='Fiyat',
    xaxis_rangeslider_visible=False,
)

fig.show()
```

<br><br><br><br>

# rsi degeriin hesaplanması

```python
veri["rsi"] = ta.rsi(veri["Close"],14)
#rsi degeri için padas_ta kütüphanesi kullandım
```

```python
plt.figure(figsize=(15,5))
plt.plot(veri["rsi"])
plt.axhline(50,color="black")
plt.axhline(30, linestyle='--',color='green', label='30 seviyesinin altındaki RSI değeri,alım sinyali olarak algılanabilir')
plt.axhline(70, linestyle='--',color='red', label='70 seviyesinin üzerindeki RSI değeri,satış sinyali olarak algılanabilir')
plt.legend()
```

<br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br>

<h1 style="color: green; text-align: center;">Alış ve Satış Stratejimiz</h1>

```python
veri["sat_sainyali"] = np.where((veri["Close"]>veri["Üst_bant"])&(veri["Close"].shift(1)>veri["Üst_bant"]) &
                                 (veri["rsi"]>veri["rsi"].shift(1)),1,0)
veri["al_sinyali"] = np.where((veri["alt_bant"]>veri["Close"])&(veri["Close"].shift(1)>veri["alt_bant"]) &
                                 (veri["rsi"]<veri["rsi"].shift(1)),1,0)
```

```python
plt.figure(figsize=(15,8))
plt.plot(veri.loc[veri["sat_sainyali"]== 1]["Close"],"v",label="Sat Sinyali",color="red")
plt.plot(veri.loc[veri["al_sinyali"]== 1]["Close"],"^",label="Al Sinyali",color="green")
plt.plot(veri["Close"])
plt.legend()
plt.grid()
#strateji sonucunda oluşan sinyaller
```

<h3 style="color:green">Öncelikle detaylı analiz için VectorBT backtest kütüphanesini kullanalım. Bu şekilde stratejimiz hakkında daha detaylı bilgi alabiliriz.</h3>

```python
import vectorbt as vbt
close = veri["Close"] #kapanış fiyatları
entries = veri["al_sinyali"].astype(bool) #1 degerleri al pozisyonları ve tarihlerini yolluyoruz
exits = veri["sat_sainyali"].astype(bool) #0 degerleri sat pozisyonu ve sinyallerini yolluyoruz

pf = vbt.Portfolio.from_signals(
    close, #kapanış degeri
    entries,
    exits,
    init_cash=1000, #başlangıç aşamasında paramız
    fees=0.001 #her işlem sonrası bizden alınacak olan komisyon
)

print(pf.stats())
```

# Stratejimizi detaylı test edelim

```python
elimizdeki_para = 1000  # TL
hisse_adeti = 0
toplam_kazanc = 0
para_grafik = []
hisse_Sayaci = []
elimizdeki_para_Sayaci = [1000]
for i in range(len(veri)):
    if veri.iloc[i]["al_sinyali"] == 1: 
        adet = int(elimizdeki_para /veri.iloc[i]["Close"])  
        hisse_adeti += adet
        elimizdeki_para -= adet * veri.iloc[i]["Close"]  
        hisse_Sayaci.append(hisse_adeti)
        elimizdeki_para_Sayaci.append(elimizdeki_para)
    elif veri.iloc[i]["sat_sainyali"]== 1: 
        elimizdeki_para += hisse_adeti * veri.iloc[i]["Close"]  
        hisse_adeti = 0  
        hisse_Sayaci.append(hisse_adeti)
        elimizdeki_para_Sayaci.append(elimizdeki_para)

    toplam_deger = elimizdeki_para + hisse_adeti * veri.iloc[i]["Close"] 
    para_grafik.append(toplam_deger) 


elimizdeki_para += hisse_adeti * veri.iloc[-1]["Close"]
hisse_adeti = 0
para_grafik.append(elimizdeki_para)  
hisse_Sayaci.append(hisse_adeti)
elimizdeki_para_Sayaci.append(elimizdeki_para)


toplam_kazanc = elimizdeki_para - 1000  

print("Toplam Kazanç:", toplam_kazanc, "₺")
print("Elimizdeki Para:", elimizdeki_para, "₺")
print("Hisse Adedi:", hisse_adeti)

```
